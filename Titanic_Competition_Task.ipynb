```python
# Any results you write to the current directory are saved as output.

# Load the training data
train_data = pd.read_csv("/content/train.csv")
# Display the first few rows of the training data
print("Training Data Head:")
print(train_data.head())
```

```python
# Load the test data
test_data = pd.read_csv("/content/test.csv")
# Display the first few rows of the test data
print("\nTest Data Head:")
print(test_data.head())
```

```python
# Calculate the survival rate of women in the training data
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print(f"\n% of women who survived: {rate_women:.2f}")
```

```python
# Calculate the survival rate of men in the training data
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print(f"% of men who survived: {rate_men:.2f}")
```

```python
# Import the RandomForestClassifier model from scikit-learn
from sklearn.ensemble import RandomForestClassifier

# Define the target variable (what we want to predict)
y = train_data["Survived"]

# Define the features (columns) to be used for prediction
features = ["Pclass", "Sex", "SibSp", "Parch"]
```

```python
# Convert categorical features (like 'Sex') into numerical format using one-hot encoding
# This is required for many machine learning models
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
# Initialize the Random Forest Classifier model
# n_estimators: The number of trees in the forest.
# max_depth: The maximum depth of the tree.
# random_state: Controls the randomness of the bootstrapping of the samples and the splitting of features.
#               Using a fixed random_state ensures reproducibility.
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)

# Train the model using the training data
model.fit(X, y)
```

```python
# Make predictions on the test data
predictions = model.predict(X_test)

# Create a Pandas DataFrame for the submission file
# It should contain 'PassengerId' from the test data and our 'Survived' predictions
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
```

```python
# Save the submission DataFrame to a CSV file
# index=False prevents Pandas from writing the DataFrame index as a column in the CSV
output.to_csv('submission.csv', index=False)

print("\nYour submission was successfully saved to submission.csv!")
```
